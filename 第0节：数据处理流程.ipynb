### 引入相关库

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 数据收集

这里只使用 'property_type', 'room_type', 'accommodates', 'bedrooms', 'price'列

```python
listing_data = pd.read_csv('./data/listings.csv', usecols=['property_type', 'room_type', 'accommodates',
                                                                             'bedrooms', 'price'])
```

```python
listing_data.head()
```

### 数据处理

price 列的 数据 转化为 数值型

```python
price_data_list = listing_data['price']
```

```python
price_data_list.head()
```

```python
price_data_list = listing_data['price'].tolist()
```

```python
print price_data_list
```

```python
print [item  for item in price_data_list]
```

```python
### 使用列表解析

price_list = [float(item.replace(',','')[1:])  for item in price_data_list]
```

```python
print price_list
```

```python
listing_data['price'] = price_list
listing_data.head()
```

### 数据分析及展现

```python
plt.figure(figsize = (12, 6))

sns.boxplot(x = 'property_type', y = 'price',  data = listing_data)

xt = plt.xticks(rotation=90)
```

```python
sns.factorplot('accommodates', 'price', data = listing_data, color = 'm', \
               estimator = np.median, size = 7.5,  aspect=1.35)
xt = plt.xticks(rotation=90)
```

```python
listing_data.groupby(['property_type', 'bedrooms'])['price'].mean().unstack()
```

```python
plt.figure(figsize=(10,10))

sns.heatmap(listing_data.groupby(['property_type', 'bedrooms'])['price'].mean().unstack(), annot=True, fmt=".0f")
```
